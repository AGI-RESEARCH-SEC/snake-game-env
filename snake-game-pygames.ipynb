```python

# source: https://www.edureka.co/blog/snake-game-with-pygame/
# note: edited
'''a simple snake game using pygame'''

import pygame
import time
import random
 
pygame.init()
 
white = (255, 255, 255)
yellow = (255, 255, 102)
black = (0, 0, 0)
red = (213, 50, 80)
green = (0, 255, 0)
blue = (50, 153, 213)
 
dis_width = 600     # width of the display
dis_height = 400    # height of the display
 
dis = pygame.display.set_mode((dis_width, dis_height))  # set the display
pygame.display.set_caption('Snake Game by Edureka')
 
clock = pygame.time.Clock()     # create a clock object
 
snake_block = 10    # size of the snake block
snake_speed = 15    # speed of the snake
 
font_style = pygame.font.SysFont("bahnschrift", 25) # font style
score_font = pygame.font.SysFont("comicsansms", 35)
 
 
def Your_score(score):
    value = score_font.render("Your Score: " + str(score), True, yellow)
    dis.blit(value, [0, 0])     # blit the value on the screen. blit means draw
 
 
 
def our_snake(snake_block, snake_list):
    for x in snake_list:
        pygame.draw.rect(dis, black, [x[0], x[1], snake_block, snake_block])    # draw a rectangular snake
 
# function to display message on screen
def message(msg, color):
    mesg = font_style.render(msg, True, color)
    dis.blit(mesg, [dis_width / 6, dis_height / 3])
 
# main function
def gameLoop():
    game_over = False
    game_close = False
 
    # initial position of the snake
    x1 = dis_width / 2
    y1 = dis_height / 2
    
    # stationary snake initially
    x1_change = 0
    y1_change = 0
 
    snake_List = []
    Length_of_snake = 1
 
    # randomly place the food
    foodx = round(random.randrange(0, dis_width - snake_block) / 10.0) * 10.0
    foody = round(random.randrange(0, dis_height - snake_block) / 10.0) * 10.0
 
    while not game_over:
    
            while game_close == True:
                dis.fill(blue)
                message("Crash! Press C-Play Again or Q-Quit", red)
                Your_score(Length_of_snake - 1)
                pygame.display.update()
    
                for event in pygame.event.get():
                    if event.type == pygame.KEYDOWN:
                        if event.key == pygame.K_q:
                            game_over = True    # quit the game on pressing q
                            game_close = False
                        if event.key == pygame.K_c:
                            gameLoop()  # restart the game on pressing c
    
            for event in pygame.event.get():
                if event.type == pygame.QUIT:
                    game_over = True
                if event.type == pygame.KEYDOWN:
                    
                    if event.key == pygame.K_LEFT:
                        # snake moving right should ignore left key press
                        if len(snake_List) == 1 or ([x1 - snake_block, y1]  not in snake_List):
                            x1_change = -snake_block
                            y1_change = 0
                    elif event.key == pygame.K_RIGHT:
                        # snake moving left should ignore right key press
                        if len(snake_List) == 1 or ([x1 + snake_block, y1]  not in snake_List):
                            x1_change = snake_block
                            y1_change = 0
                    elif event.key == pygame.K_UP:
                        # snake moving down should ignore up key press
                        if len(snake_List) == 1 or ([x1, y1 - snake_block] not in snake_List):
                            y1_change = -snake_block
                            x1_change = 0
                    elif event.key == pygame.K_DOWN:
                        # snake moving up should ignore down key press
                        if len(snake_List) == 1 or ([x1, y1 + snake_block]  not in snake_List):
                            y1_change = snake_block
                            x1_change = 0
                    # pause the game on pressing p
                    print(f'\nsnake_block: {snake_List}, next:[{x1 + x1_change}, {y1 + y1_change}]')

                        
    
            # snake goes out of the display
            if x1 >= dis_width or x1 < 0 or y1 >= dis_height or y1 < 0:
                game_close = True
            
            
            x1 += x1_change
            y1 += y1_change

            
            dis.fill(blue)  # background color
            pygame.draw.rect(dis, green, [foodx, foody, snake_block, snake_block])  # draw the food
            
            snake_Head = []
            snake_Head.append(x1)
            snake_Head.append(y1)
            snake_List.append(snake_Head)
            
            # collision with itself
            if len(snake_List) > Length_of_snake:
                del snake_List[0]
    
            for x in snake_List[:-1]:
                if x == snake_Head:
                    game_close = True
    
            our_snake(snake_block, snake_List)
            Your_score(Length_of_snake - 1)
    
            pygame.display.update()
    
            if x1 == foodx and y1 == foody:
                foodx = round(random.randrange(0, dis_width - snake_block) / 10.0) * 10.0
                foody = round(random.randrange(0, dis_height - snake_block) / 10.0) * 10.0
                Length_of_snake += 1
    
            clock.tick(snake_speed)
    
        
 
 
gameLoop()
```
